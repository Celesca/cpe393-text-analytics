# 1. Loading data

```python
import nltk
# nltk.download('all')
# nltk.download() #for the missing library
f = open('books.txt', encoding='utf-8')
raw = f.read()
tokens = nltk.word_tokenize(raw)
```

```python
print(tokens[1000:1100])
```

# 2. Regular expression

```python
tokens = [w.lower() for w in tokens]
```

```python
import re

[w for w in tokens if re.search('ed$', w)]
```

# 3. Regular expression tokenizer

## 3.1 Basic regex tokenizer

```python
s = ("Good muffins cost $3.88\nin New York.  Please buy me\n"
      "two of them.\n\nThanks.")
s2 = ("Alas, it has not rained today. When, do you think, "
      "will it rain again?")
s3 = ("<p>Although this is <b>not</b> the case here, we must "
      "not relax our vigilance!</p>")
```

```python
s2
```

```python
nltk.regexp_tokenize(s2, r'[,\.\?!"]\s*', gaps=True)
```

What happened after applying the above RegEx tokenizer to the input string?

Ans: เมื่อทำการ Tokenize จะได้เป็น List ของประโยค และทำการกำจัดตัวอักษรที่ตัดคำ หรือก็คือตัวเครื่องหมายสัญลักษณ์ที่ไม่ใช่ตัวอักษรเช่น , . ? !

```python
s3
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=False)
```

What happened after applying the above RegEx tokenizer to the input string?

Ans: เมื่อทำการ Tokenize โดย gaps = False จะได้เป็นคำที่เป็น Tag ของ HTML ที่มีปีกกา </> ด้านใน Tag จะเป็นคำใดก็ได้

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=True)
```

What happened after applying the above RegEx tokenizer to the input string?

Ans: เมื่อทำการ Tokenize โดย gaps = True จะได้เป็นคำที่มาจากการกำจัด Tag HTML ที่คั่นระหว่างประโยคหรือคำอยู่

#### 3. HTML Hyperlink patterns

```python
import re

link_pattern = "(^<a.*>(.*?)</a.*>$)|(^<a.*/>$)"

tokens = ['<a href=""></a>', 
          '<a href="https://www.google.com">Click here</a>', 
          '<a href="https://www.facebook.com"/>', 
          '<a data-aos="fade-up" href="https://www.twitter.com"/>', 
          '<a className="text-center" href="https://www.instagram.com">Click the link please!</a>', 
          '<a href="https://www.linkedin.com">']

[w for w in tokens if re.search(link_pattern, w)]
```

* Why you want to discover those patterns

* Why are they interesting to you?

  * I learned about SEO and the concept of "Backlinks". I knew that there are Google’s robots check all the hyperlinks on a website. That’s why I’m interested in hyperlink patterns.
